# Remove Files without Transcription

```python
import glob
files = glob.glob("/Volumes/GoogleDrive/My Drive/aeneas/job/OEBPS/Resources/*.txt")
move_to = "/Volumes/GoogleDrive/My Drive/stress"
```

```python
remove = []
for f in files:
    fo = open(f, "r")
    lines = fo.readlines()
if not any([bool(not s or s.isspace()) for s in lines]):
        remove.append(f)
```

```python
import shutil
import os
other = "/Volumes/GoogleDrive/My Drive/stress"
files = glob.glob("/Volumes/GoogleDrive/My Drive/aeneas/job/OEBPS/Resources/*.txt")

for text_fn in remove:
    if text_fn not in files:
        continue
    youtube_id = os.path.basename(text_fn)[:-4]
    audio_fn = "/Volumes/GoogleDrive/My Drive/aeneas/job/OEBPS/Resources/{}.wav".format(youtube_id)
    output_txt = other + '/{}.txt'
    output_wav = other + '/{}.wav'
    output_txt = output_txt.format(youtube_id)
    output_wav = output_wav.format(youtube_id)
    shutil.move(text_fn,output_txt)
    shutil.move(audio_fn,output_wav)
    print("moved to {} and {}".format(output_txt, output_wav))
```

# Prepare files for aeneas batch processing

```python
count = 1

config = """is_hierarchy_type=flat
is_hierarchy_prefix=OEBPS/Resources/
is_text_file_relative_path=.
is_text_file_name_regex=.*\.txt
is_text_type=plain
is_audio_file_relative_path=.
is_audio_file_name_regex=.*\.wav

os_job_file_name=output_example{}
os_job_file_container=zip
os_job_file_hierarchy_type=flat
os_job_file_hierarchy_prefix=OEBPS/Resources/
os_task_file_name=$PREFIX.csv
os_task_file_format=csv

job_language=en"""
config_path = "/Volumes/GoogleDrive/My Drive/aeneas/jobs/job{}/config.txt"
target = "/Volumes/GoogleDrive/My Drive/aeneas/jobs/job{}/OEBPS/Resources/{}"

allfiles = glob.glob("/Volumes/GoogleDrive/My Drive/aeneas/job/OEBPS/Resources/*.txt")
for i, text_fn in enumerate(allfiles):
    youtube_id = os.path.basename(text_fn)[:-4]
    audio_fn = "/Volumes/GoogleDrive/My Drive/aeneas/job/OEBPS/Resources/{}.wav".format(youtube_id)
    if not os.path.exists(target.format(count, '')):
        os.makedirs(target.format(count, ''))
    shutil.move(text_fn,target.format(count, youtube_id + '.txt'))
    shutil.move(audio_fn,target.format(count, youtube_id + '.wav'))
    
    if i %100 == 0 and i != 0:
        with open(config_path.format(count), 'w') as out_file:
             out_file.write(config.format(count))
        count += 1
```
